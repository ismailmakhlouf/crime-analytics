```
# TODO
from pyspark.sql.functions import col, lower, month
robberyLosAngelesDF = (crimeDataLosAngelesDF
                      .select(month("timeOccurred").alias("month"),col("crimeCodeDescription").alias("robberies"))
                       .filter(lower(col("crimeCodeDescription")) == "robbery")
                      )
```
